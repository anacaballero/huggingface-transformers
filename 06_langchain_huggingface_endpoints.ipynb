## Prompt templates and chaining
We'll use two of the core components in LangChain: **prompt templates** and **chains**!
- Prompt templates are used for creating prompts in a more modular way, so they can be reused and built on.
- Chains act as the glue in LangChain; bringing the other components together into workflows that pass inputs and outputs between the different components.

We need the classes HuggingFaceEndpoint (they have to be pre-loaded).

`````python
from langchain import PromptTemplate, LLMChain
`````

`````python
# Create a prompt template from the template string
template = "You are an artificial intelligence assistant, answer the question. {question}"
prompt = PromptTemplate(template=template, input_variables=['question'])

# Create a chain to integrate the prompt template and LLM
llm = HuggingFaceEndpoint(repo_id='tiiuae/falcon-7b-instruct', huggingfacehub_api_token=huggingfacehub_api_token)
llm_chain = prompt | llm

question = "How does LangChain make LLM application development easier?"
print(llm_chain.invoke({"question": question}))
`````

## Managing Chat Model Memory
### Integrating a chatbot message history
A key feature of chatbot applications is the ability to have a conversation, where context from the conversation history is stored and available for the model to access.




```python
from langchain.memory
# Set your API Key from OpenAI
openai_api_key= '____'
llm = ChatOpenAI(temperature=0, openai_api_key=openai_api_key)

# Create the conversation history and add the first AI message
history = ChatMessageHistory
history.sdd_ai_message("Hello! Ask me anything about Python programming!")

# Add the user message to the history
history.add_user_message("What is a list comprehension?")

# Add another user message and call the model
history.add_user_message("Describe the same in fewer words")
response = llm.invoke(history.messages)
print(response.content)
```

### Creating a memory buffer
For many applications, storing and accessing the entire conversation history isn't technically feasible. In these cases, the messages must be condensed while retaining as much relevant context as possible. One common way of doing this is with a memory buffer, which stores only the most recent messages.



### Instructions
* Assign your OpenAI API key to openai_api_key.
* Define a buffer memory that stores the four most recent messages.
* Define a conversation chain for integrating the model and memory buffer.
* Invoke the chain twice with the inputs provided.

```python
from langchain.memory import ConversationBufferMemory
from langchain.chains import ConversationChain


# Set your API Key from OpenAI
openai_api_key = '____'
chat = OpenAI(model_name="gpt-3.5-turbo-instruct", temperature=0, openai_api_key=openai_api_key)

llm= ChatOpenAI(openai_api_key=openai_api_key)

# Define a buffer memory
memory = ConversationBufferMemory(size=4)

# Define the chain for integrating the memory with the model
buffer_chain = ConversationChain(lln=llm, memory=memory)

# Invoke the chain with the inputs provided
buffer_chain.invoke("Write Python code to draw a scatter plot.")
buffer_chain.invoke("Use the Seaborn library.")
```

### **Outline: Managing Memory in Chat Models**

**Objective**: Improve the model's ability to follow conversations, ask follow-up questions, and adapt to the user.

#### **1. Memory in Conversation**
- **Importance**:
  - Enables follow-up questions and iterative responses.
  - Helps adapt to user preferences.
  - Limited by the model's **context window** (the amount of text the model can process at once).
- **LangChain Classes for Memory**:
  - `ChatMessageHistory`
  - `ConversationBufferMemory`
  - `ConversationSummaryMemory`

#### **2. ChatMessageHistory**
- **Function**: Stores the entire history of messages between the user and the model.
- **Implementation**:
  - Import `ChatMessageHistory` and `ChatOpenAI`.
  - Start the conversation history with `ChatMessageHistory`.
  - Add AI messages with `.add_ai_message()`.
  - Add user messages with `.add_user_message()`.
  - Use the model on the `messages` attribute to generate responses (`.content`).

#### **3. Updating ChatMessageHistory**
- **Use**: Responses based on the full context stored in the conversation history.

```python
from langchain.memory import ChatMessageHistory
from langchain_openai import ChatOpenAI

# Set your API Key from OpenAI
openai_api_key= '----My_OpenAI_key-------'
llm = ChatOpenAI(temperature=0, openai_api_key=openai_api_key)

# Create the conversation history and add the first AI message
history = ChatMessageHistory
history.sdd_ai_message("Hello! Ask me anything about Python programming!")

# Add the user message to the history
history.add_user_message("What is a list comprehension?")

# Add another user message and call the model
history.add_user_message("Describe the same in fewer words") #Updating
response = llm.invoke(history.messages)
print(response.content)
```
#### **4. ConversationBufferMemory**
- **Function**: Maintains a buffer of the last few messages in the conversation.
- **Implementation**:
  - Specify the number of messages to store (`size`).
  - Integrate with `ConversationChain`.

#### **5. Using ConversationBufferMemory**
- **Example**:
  - Pass a series of inputs, with the last one asking the model to recall the first question.
  - The chain returns a dictionary with the final input, the history, and the model's response.

```python

from langchain.chains import ConversationChain
```

#### **6. ConversationSummaryMemory**
- **Function**: Summarizes the conversation over time, retaining key information.
- **Implementation**:
  - Use `ConversationChain`.
  - `ConversationSummaryMemory` requires an LLM to create summaries.
  - Each message triggers a separate LLM call to summarize the history.
  - Set `verbose=True` to see the model's decisions.

#### **7. Applying ConversationSummaryMemory**
- **Example**:
  - Ask the model three sequential questions.
  - The output shows a summary of the first two inputs and responses, which is used to answer the final input.

```python
from langchain.memory import ConversationSummaryMemory

llm=ChatOpenAI(openai_api_key=openai_api_key)

memory = ConversationSummaryMemory(llm=ChatOpenAI(openai_api_key=openai_api_key))

summary_chain=ConversationChain(llm=llm, memory=memory, verbose=True)

summary_chain.invoke("Please summarize the future in 2 sentences")
summary_chain.invoke("Why?")
summary_chain.invoke("What will I need to shape this?") #¿Qué necesitaré para dearle forma a esto?
````
